```python
import numpy as np
import matplotlib.pyplot as plt
def bounded_walk(boundary=20):
    pos = 0
    while abs(pos) < boundary:
        pos += np.random.choice([-1, 1])
    return pos

results = [bounded_walk() for _ in range(10000)]
print(f"+20: {np.mean(np.array(results) > 0):.3f}, -20: {np.mean(np.array(results) < 0):.3f}")
```

```python
def walk_2d(n=2000):
    moves = np.random.randint(0, 4, n)
    dx = np.array([0,0,1,-1])[moves]
    dy = np.array([1,-1,0,0])[moves]
    return np.concatenate([[0], np.cumsum(dx)]), np.concatenate([[0], np.cumsum(dy)])

for _ in range(5):
    x, y = walk_2d()
    plt.plot(x, y, alpha=0.7, linewidth=0.8)
plt.axis('equal'); plt.show()
```

```python
def episode(start):
    s = start
    while s not in (0, 6):
        s += np.random.choice([-1, 1])
    return 1.0 if s == 6 else 0.0

for s, name in zip([1,2,3,4,5], 'ABCDE'):
    v = np.mean([episode(s) for _ in range(50000)])
    print(f"{name}: empirical={v:.3f}, true={s/6:.3f}")
```
