Q1 
Change mu and sigma in section 2. What happens to the histogram shape?

```python

import numpy as np
import matplotlib.pyplot as plt

for mu, sigma, label in [(0, 1, 'mu=0, sigma=1'), (5, 1, 'mu=5, sigma=1'), (0, 3, 'mu=0, sigma=3')]:
    s = np.random.normal(mu, sigma, 10000)
    plt.hist(s, bins=50, alpha=0.5, density=True, label=label)
plt.legend(); plt.show()
```

Q2 In section 4, replace dice rolls with samples from an exponential distribution (np.random.exponential). Does LLN still hold?

```python
lam = 0.5
samples = np.random.exponential(1/lam, 5000)
running_mean = np.cumsum(samples) / np.arange(1, 5001)
plt.plot(running_mean, label='Running mean')
plt.axhline(1/lam, color='red', linestyle='--', label='True mean = 2.0')
plt.legend(); plt.show()
```

Q3 Challenge: simulate 1000 coin flips (0/1). Plot the running fraction of heads. What does it converge to?

```python
flips = np.random.randint(0, 2, 1000)
running = np.cumsum(flips) / np.arange(1, 1001)
plt.plot(running); plt.axhline(0.5, color='red', linestyle='--'); plt.show()
```
